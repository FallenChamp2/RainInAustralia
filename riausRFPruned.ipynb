```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
k = RIAUS['WindGustDir']
RIAUS['WindGustDir'] = encoder.fit_transform(k)
```

```python
k = RIAUS['WindDir9am']
RIAUS['WindDir9am'] = encoder.fit_transform(k)

k = RIAUS['WindDir3pm']
RIAUS['WindDir3pm'] = encoder.fit_transform(k)

k = RIAUS['RainToday']
RIAUS['RainToday'] = encoder.fit_transform(k)

k = RIAUS['RainTomorrow']
RIAUS['RainTomorrow'] = encoder.fit_transform(k)

k = RIAUS['Location']
RIAUS['Location'] = encoder.fit_transform(k)
```

```python
RIAUS.head()
```

```python
Date = pd.to_datetime(RIAUS['Date'], infer_datetime_format = True) 
```

```python
X = RIAUS.iloc[ : , 1 : 22]
```

```python
#X = X[ : 2000]
```

```python
X['Evap - Sun'] = X['Evaporation'] - X['Sunshine']
```

```python
X['WS * Cloud3pm * Sunshine * MinTemp'] = X['WindSpeed3pm'] * X['Sunshine'] * X['Cloud3pm'] * X['MinTemp']
```

```python
X['MaxT - MinT'] = X['MaxTemp'] - X['MinTemp']
```

```python
X['WindDir'] = X['WindDir3pm'] - X['WindDir9am']
```

```python
X['WindSpeed'] = X['WindSpeed3pm'] - X['WindSpeed9am']
```

```python
X['Humidity'] = X['Humidity3pm'] - X['Humidity9am']
```

```python
X['Pressure'] = X['Pressure3pm'] - X['Pressure9am']
```

```python
X['Cloud'] = X['Cloud3pm'] - X['Cloud9am']
```

```python
X['Temp'] = X['Temp3pm'] - X['Temp9am']
```

```python
X.head()
```

```python
X['Month'] = Date.dt.month
```

```python
X['Year'] = Date.dt.year
```

```python
K = X
```

```python
y = RIAUS['RainTomorrow']
```

```python
#y = y[ : 2000]
```

```python
K.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(K)
K = scaled_data
```

```python
from sklearn.decomposition import PCA

pca2 = PCA(n_components = 10)
components2 = pca2.fit_transform(K)
components2_df = pd.DataFrame(components2)
```

```python
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

model = DecisionTreeClassifier()

alphas = model.cost_complexity_pruning_path(components2_df, y)['ccp_alphas']
alphas
```

parameters = {
            'ccp_alpha' : alphas
}

modelRSCV = RandomizedSearchCV(estimator = model, param_distributions = parameters, n_iter = 10, cv = 5, verbose = 3,
                              n_jobs = -1, scoring = 'accuracy')

modelRSCV.fit(components2_df, y)

print('Best params', modelRSCV.best_params_)
print('Best estimator', modelRSCV.best_estimator_)
print('Best score', modelRSCV.best_score_)

```python
model2 = RandomForestClassifier(n_estimators = 1000, max_features = 'sqrt', max_depth = 15, min_samples_split = 10,
                               min_samples_leaf = 2, bootstrap = True, ccp_alpha = 0.003533)

score = cross_val_score(model2, components2_df, y, cv = 5)
score.mean()
```
